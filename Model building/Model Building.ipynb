```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

df = pd.read_csv("new heart dataframe.csv")
```

```python
df = pd.get_dummies(df, columns=["Thal"])
df["Thal_fixed"] = df["Thal_fixed"].astype(int)
df["Thal_reversable"] = df["Thal_reversable"].astype(int)
df["Thal_normal"] = df["Thal_normal"].astype(int)
```

I did one-hot encoding on the "Thal" column above

```python
typical1 = len(df[(df["Chest pain"] == "typical") & (df["Target"] == 1)])
print(typical1)
typical0 = len(df[(df["Chest pain"] == "typical") & (df["Target"] == 0)])
print(typical0)
```

```python
nontypical1 = len(df[(df["Chest pain"] == "nontypical") & (df["Target"] == 1)])
print(nontypical1)
nontypical0 = len(df[(df["Chest pain"] == "nontypical") & (df["Target"] == 0)])
print(nontypical0)
```

```python
asymptomatic1 = len(df[(df["Chest pain"] == "asymptomatic") & (df["Target"] == 1)])
print(asymptomatic1)
asymptomatic0 = len(df[(df["Chest pain"] == "asymptomatic") & (df["Target"] == 0)])
print(asymptomatic0)
```

```python
nonanginal1 = len(df[(df["Chest pain"] == "nonanginal") & (df["Target"] == 1)])
print(nonanginal1)
nonanginal0 = len(df[(df["Chest pain"] == "nonanginal") & (df["Target"] == 0)])
print(nonanginal0)
```

```python
typical_mean =  (23*(7/23) + 25*(139/303)) / (23+25)
nontypical_mean = (50*(9/50) + 25*(139/303)) / (50+25)
asymptomatic_mean = (144*(105/144) + 25*(139/303)) / (144+25)
nonanginal_mean = (86*(18/86) + 25*(139/303)) / (86+25)
# weighted means to pull data up-down towards the mean and avoid noise
print(typical_mean)
print(nontypical_mean)
print(asymptomatic_mean)
print(nonanginal_mean)
```

```python
df.loc[df["Chest pain"] == "typical", "Chest pain"] = typical_mean
df.loc[df["Chest pain"] == "nontypical", "Chest pain"] = nontypical_mean
df.loc[df["Chest pain"] == "asymptomatic", "Chest pain"] = asymptomatic_mean
df.loc[df["Chest pain"] == "nonanginal", "Chest pain"] = nonanginal_mean
```
