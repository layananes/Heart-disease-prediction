```python
columns = df.select_dtypes(include=['number']).columns

for column in columns:
    plt.figure(figsize=(8, 3)) # Adjust figure size for compactness
    sns.boxplot(x=df[column])
    plt.title(f'Outliers in: {column}')
    plt.show()
```

```python
columns = ['Rest BP', 'Chol', 'Max HR', 'Old Peak']
for column in columns:
    print("outliers for: ", column)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    LowerBound = Q1 - 1.5 * IQR
    UpperBound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < LowerBound) | (df[column] > UpperBound)]
    print(outliers)
```

```python
columns = ['Rest BP', 'Chol', 'Max HR', 'Old Peak']
for column in columns:
    print( column)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    LowerBound = Q1 - 1.5 * IQR
    UpperBound = Q3 + 1.5 * IQR
    outliers = (df[column] < LowerBound) | (df[column] > UpperBound)
    median = df[column].median()
    df.loc[outliers, column] = median
    print(LowerBound)
    print(UpperBound)



# this loop replaces each outlier in each column
# with the median for each column
```
