```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv("new heart dataframe.csv")
columns = ['Rest BP', 'Chol', 'Max HR', 'Old Peak']
for column in columns:
    plt.figure(figsize=(8, 3)) # Adjust figure size for compactness
    sns.boxplot(x=df[column])
    plt.title(f'Outliers in: {column}')
    plt.show()

print(df)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
# Calcium, Old peak, and Ex Ang show a good correlation with heart disease.
```

```python
sns.histplot(df["Rest BP"]) 
print(df["Rest BP"].skew())
# slightly right scewed, though it can generally be said that the values are symmetrical
# with a big chunk of the values concentrated around 120 - 140 which is generally normal
```

```python
sns.histplot(df["Chol"]) 
print(df["Chol"].skew())
# slightly right skewed. generally symmetrical, most of the being around 200 - 270 
# mainly normal (except for the values above 240)
```

```python
sns.histplot(df["Max HR"]) 
print(df["Max HR"].skew())
# left skewed. more data is concentrated at the right around 150 - 170 which is normal-
```

```python
sns.histplot(df["Old Peak"]) 
print(df["Old Peak"].skew())

# clearly right skewed! most data concentrated at around 0 - 0.5, meaning the drop
# is not large and is normal
```

```python
sns.histplot(df["Slope"]) 
print(df["Slope"].skew())
# propable right skew. Most values are at 1 and 2. 1 is healthy, while 2 and above 
# is dangerous. it makes sense that both histograms are the same height, since more than
# half of the patients had heart disease.
```

```python
sns.histplot(df["Ca"]) 
print(df["Ca"].skew())
# right scewed. Higher calcium usually signals higher chance of heart disease, so
# the distribution makes sense
```
