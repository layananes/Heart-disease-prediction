```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv("new heart dataframe.csv")
columns = ['Rest BP', 'Chol', 'Max HR', 'Old Peak']
for column in columns:
    plt.figure(figsize=(8, 3)) # Adjust figure size for compactness
    sns.boxplot(x=df[column])
    plt.title(f'Outliers in: {column}')
    plt.show()
```
